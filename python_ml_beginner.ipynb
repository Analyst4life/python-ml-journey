```python
#1. Classification Example: Predicting Iris Species
```

```python
#Classification predicts which category an observation belongs to.
#The code below utilizes the Iris dataset to determine a flower’s species by analyzing its petal and sepal dimensions.
#Import necessary libraries.
```

```python
#The above introduces machine learning in Python, showing how to use Scikit-learn for tasks like classification, regression, and clustering, with an example that predicts a flower’s species from the Iris dataset and begins by importing the needed libraries.
```

```python
from sklearn.datasets import load_iris # Load sample dataset
```

```python
from sklearn.model_selection import train_test_split # Split dataset into training and test sets
```

```python
from sklearn.neighbors import KNeighborsClassifier # K-Nearest Neighbors classifier
```

```python
# Split the dataset into training and testing sets
# x_train, y_train: features and labels used to train the model
# x_test, y_test: features and labels used to evaluate the model's performance
# test_size=0.2 means 20% of the data is reserved for testing, 80% for training
# random_state=42 ensures the split is reproducible (same split every time)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#The above code splits the data into two parts: one part (80%) is used to train the model, and the other part (20%) is used to test how well the model performs. The random_state=42 ensures that the split is the same every time you run the code.
```

```python
# Initialize the K-Nearest Neighbors classifier
knn = KNeighborsClassifier(n_neighbors=3)  # 3 nearest neighbors
```

```python
# Train the model
knn.fit(x_train, y_train)
```

```python
# Predict the species for test data
y_pred = knn.predict(x_test)

# Evaluate accuracy
accuracy = knn.score(x_test, y_test)  # Returns accuracy as a percentage
print("Classification Accuracy:", accuracy)
```

```python
#This code shows a simple example of machine learning using the Iris flower dataset. We use the measurements of flowers to teach a model how to recognize their species. First, the data is split into two parts: one to train the model and one to test it. Then we use a K-Nearest Neighbors (KNN) classifier to learn from the training data. Finally, the model predicts the species of the test flowers and checks how accurate its predictions are.
```

```python
#2. Regression Example: Predicting House Prices
```

```python
#Regression is used to predict a continuous value. In this example, a dataset of house sizes is used to estimate house prices.
```

```python
#Import the NumPy library for numerical operations and the LinearRegression model from Scikit-learn to perform linear regression.
```

```python
# Import libraries
import numpy as np
from sklearn.linear_model import LinearRegression  # Linear regression model
```

```python
#Set up the data for a prediction. 
```

```python
# Sample data: house sizes (in square feet) and prices (in $1000s)
x = np.array([600, 800, 1000, 1200, 1400]).reshape(-1, 1)  # Feature: house size
y = np.array([150, 200, 250, 300, 350])  # Target: price
```

```python
#"x" holds the house sizes as input values, and y holds the corresponding house prices as output values. It represents the relationship between house size and price for prediction.
```

```python
# Initialize the Linear Regression model
model = LinearRegression()
```

```python
# Train the model to learn the relationship between x and y.
model.fit(x, y)
```

```python
# Predict the price of a new house
new_house_size = np.array([[1100]])  # Size of new house
predicted_price = model.predict(new_house_size)
```

```python
#The above code predicts the price of a house with size 1100 using the trained model.
```

```python
print("Predicted Price for 1100 sq ft house:", predicted_price[0]) # This prints the predicted price for a house with 1100 sq ft.
```

```python
#3. Clustering Example: Grouping Customers
```

```python
#Clustering: Finds patterns in unlabeled data (grouping customers).
```

```python
#This example groups customers based on annual income and spending score.
```

```python
# Import libraries
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans  # K-Means clustering
```

```python
#This code loads Matplotlib to make charts and K-Means from Scikit-learn to automatically group data into clusters for analysis and visualization.
```

```python
# Sample customer data: [Annual Income (k$), Spending Score (1-100)]
X = np.array([
    [15, 39],
    [16, 81],
    [17, 6],
    [18, 77],
    [19, 40],
    [20, 76],
    [21, 6],
    [22, 94],
    [23, 3],
    [24, 72]
])
```

```python
#This code creates a small dataset of 10 customers, where each customer is represented by two numbers: their annual income (in thousands of dollars) and their spending score (from 1 to 100). It stores the data as a NumPy array so it can be used for analysis or machine learning.
```

```python
# Initialize K-Means with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
The above code sets up the K-Means clustering model to divide the data into 3 groups (clusters). The random_state=42 ensures that the results are consistent every time you run the code.
```

```python
# Fit the model and predict clusters
clusters = kmeans.fit_predict(X)
```

```python
The above code trains the K-Means algorithm on the dataset and then assigns each data point to a cluster. It simultaneously learns where the cluster centers should be and returns an array indicating which cluster each point belongs to.
```

```python
# Plot the clusters
plt.scatter(X[:,0], X[:,1], c=clusters, cmap='viridis')  # Color points by cluster
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title("Customer Clustering")
plt.show()
```

```python
The above code creates a simple scatter plot that shows how the customers are grouped into clusters. Each point on the chart represents a customer, plotted using annual income on the x-axis and spending score on the y-axis. The points are colored based on the cluster assigned by the K-Means algorithm, making the groups easy to see. Labels and a title are added to help explain what the chart represents, and the plot is then displayed on the screen.
```

```python
The resulting plot clearly shows how the data naturally forms three distinct groups, making patterns in customer behavior easier to understand.
```
